1.Desenvolva o código para o diagrama de classes na figura abaixo, usando o padrão de design “Factory
Method”.

```python
from abc import ABC, abstractmethod

class MotorVehicle(ABC):
    @abstractmethod
    def build(self):
        pass

class MotorVehicleFactory(ABC):
    def create(self):
        vehicle = self.createMotorVehicle()
        vehicle.build()
        return vehicle

    @abstractmethod
    def createMotorVehicle(self):
        pass

class MotorcycleFactory(MotorVehicleFactory):
    def createMotorVehicle(self):
        return Motorcycle()

class CarFactory(MotorVehicleFactory):
    def createMotorVehicle(self):
        return Car()

class Motorcycle(MotorVehicle):
    def build(self):
        print("Building Motorcycle")

class Car(MotorVehicle):
    def build(self):
        print("Building Car")

if __name__ == "__main__":
    motorcycle_factory = MotorcycleFactory()
    motorcycle = motorcycle_factory.create()

    car_factory = CarFactory()
    car = car_factory.create()
```

2. Desenvolva o código para o diagrama de classes na figura abaixo, usando o padrão de design “Factory
Method”.

```python
from abc import ABC, abstractmethod

class Armchair(ABC):
    @abstractmethod
    def create(self):
        pass

class CoffeeTable(ABC):
    @abstractmethod
    def create(self):
        pass

class Sofa(ABC):
    @abstractmethod
    def create(self):
        pass

class FurnitureFactory(ABC):
    @abstractmethod
    def create_armchair(self) -> Armchair:
        pass

    @abstractmethod
    def create_coffee_table(self) -> CoffeeTable:
        pass

    @abstractmethod
    def create_sofa(self) -> Sofa:
        pass

class ModernFurnitureFactory(FurnitureFactory):
    def create_armchair(self) -> Armchair:
        return ModernArmchair()

    def create_coffee_table(self) -> CoffeeTable:
        return ModernCoffeeTable()

    def create_sofa(self) -> Sofa:
        return ModernSofa()

class RetroFurnitureFactory(FurnitureFactory):
    def create_armchair(self) -> Armchair:
        return RetroArmchair()

    def create_coffee_table(self) -> CoffeeTable:
        return RetroCoffeeTable()

    def create_sofa(self) -> Sofa:
        return RetroSofa()

class ModernArmchair(Armchair):
    def create(self):
        print("Creating Modern Armchair")

class ModernCoffeeTable(CoffeeTable):
    def create(self):
        print("Creating Modern Coffee Table")

class ModernSofa(Sofa):
    def create(self):
        print("Creating Modern Sofa")

class RetroArmchair(Armchair):
    def create(self):
        print("Creating Retro Armchair")

class RetroCoffeeTable(CoffeeTable):
    def create(self):
        print("Creating Retro Coffee Table")

class RetroSofa(Sofa):
    def create(self):
        print("Creating Retro Sofa")

if __name__ == "__main__":
    modern_factory = ModernFurnitureFactory()
    retro_factory = RetroFurnitureFactory()

    modern_armchair = modern_factory.create_armchair()
    modern_armchair.create()

    modern_coffee_table = modern_factory.create_coffee_table()
    modern_coffee_table.create()

    modern_sofa = modern_factory.create_sofa()
    modern_sofa.create()

    retro_armchair = retro_factory.create_armchair()
    retro_armchair.create()

    retro_coffee_table = retro_factory.create_coffee_table()
    retro_coffee_table.create()

    retro_sofa = retro_factory.create_sofa()
    retro_sofa.create()
```
