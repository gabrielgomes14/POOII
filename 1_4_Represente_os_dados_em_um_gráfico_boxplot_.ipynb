1. O número de pontos marcados por Dwyane Wade nos primeiros 12
jogos da temporada 2012-2013 da NBA está representado abaixo.
(Fonte: National Basketball Association.)

29 15 14 22 22 8
19 6 28 18 19 34

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

pontos = [29, 15, 14, 22, 22, 8, 19, 6, 28, 18, 19, 34]

dados = np.array(sorted(pontos))

print("Dados ordenados:", dados)
```

1.1. Encontre a média, a mediana e a moda do conjunto de dados. Qual medida
melhor representa o centro dos dados?

```python
media = np.mean(dados)

mediana = np.median(dados)

moda = stats.mode(dados)

print(f"Média: {media:.2f}")
print(f"Mediana: {mediana}")
# O conjunto de dados é bimodal (dois valores aparecem com mais frequência)
print(f"Moda(s): 19 e 22")
print(f'A média (19.5) e a mediana (19.0) estão muito próximas. Isso indica que a distribuição dos dados é razoavelmente simétrica e não possui valores extremos (outliers) que distorçam a média. Nesse caso, tanto a média quanto a mediana são boas representantes do centro dos dados. A mediana, por ser o valor central exato (19), pode ser ligeiramente preferível por não ser afetada pelos valores mais altos (34) ou mais baixos (6).')
```

1.2. Encontre a amplitude total, a variância e desvio padrão do conjunto de
dados.

```python
amplitude_total = np.max(dados) - np.min(dados)

variancia = np.var(dados, ddof=1)

desvio_padrao = np.std(dados, ddof=1)

print(f"Amplitude Total: {amplitude_total}")
print(f"Variância (amostral): {variancia:.2f}")
print(f"Desvio Padrão (amostral): {desvio_padrao:.2f}")
```

1.3. Encontre o Q1, Q2, Q3 e a amplitude interquartil (AIQ).

```python
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)

aiq = q3 - q1

print(f"Primeiro Quartil (Q1): {q1}")
print(f"Segundo Quartil (Q2 - Mediana): {q2}")
print(f"Terceiro Quartil (Q3): {q3}")
print(f"Amplitude Interquartil (AIQ): {aiq}")
```

1.4. Represente os dados em um gráfico boxplot.

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.boxplot(x=dados, color='skyblue', width=0.4)

plt.text(q1, 0.1, f'Q1: {q1}', ha='center', va='bottom', fontweight='bold')
plt.text(mediana, -0.1, f'Mediana (Q2): {mediana}', ha='center', va='bottom', fontweight='bold')
plt.text(q3, 0.1, f'Q3: {q3}', ha='center', va='bottom', fontweight='bold')
plt.text(np.min(dados), 0.1, f'Mín: {np.min(dados)}', ha='center', va='bottom', fontweight='bold')
plt.text(np.max(dados), 0.1, f'Máx: {np.max(dados)}', ha='center', va='bottom', fontweight='bold')

plt.title('Boxplot da Pontuação de Dwyane Wade (12 primeiros jogos de 2012-2013)', fontsize=16)
plt.xlabel('Pontos Marcados', fontsize=12)
plt.yticks([])

plt.show()
```
