# **Exercício 1: Leia o lucro total (total_profit) de todos os meses e mostre-o usando um gráfico de linhas**

Dados de lucro total fornecidos para cada mês. O gráfico de linha gerado deve incluir as seguintes propriedades:

*   **Label Y = lucro total**

# **Exercício 5:  Leia os dados de vendas dos cremes faciais e produtos de lavagem facial e mostre-os usando o gráfico de barras**

**O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico.**




```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("/content/company_sales_data - company_sales_data.csv")

meses = df['month_number']
vendas_creme_facial = df['facecream']
vendas_lavagem_facial = df['facewash']

x_pos = np.arange(len(meses))
largura_barra = 0.35

plt.bar(x_pos - largura_barra/2, vendas_creme_facial, width=largura_barra, label='Creme Facial')
plt.bar(x_pos + largura_barra/2, vendas_lavagem_facial, width=largura_barra, label='Lavagem Facial')
plt.xlabel('Número do mês')
plt.ylabel('Número de unidades vendidas')
plt.title('Comparativo de Vendas: Creme vs. Lavagem Facial')

plt.xticks(ticks=x_pos, labels=meses)

plt.grid(True, linestyle='--')
plt.legend()

plt.show()
```

# **Exercício 8: Calcule os dados de vendas totais do ano passado para cada produto e mostre-os usando um gráfico de pizza**





```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/company_sales_data - company_sales_data.csv")

labels = ['Creme Facial', 'Lavagem Facial', 'Pasta de Dente', 'Sabonete', 'Shampoo', 'Hidratante']

vendas_totais = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['toothpaste'].sum(),
    df['bathingsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()
]

plt.pie(vendas_totais, labels=labels, autopct='%1.1f%%')

plt.axis('equal')

plt.title('Participação de Vendas por Produto no Ano')

plt.show()
```

# **Exercício 9: Leia os dados da Lavagem Facial e do Sabonete de todos os meses e exiba-o usando o SubPlot**





```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/company_sales_data - company_sales_data.csv")

meses = df['month_number']
vendas_lavagem_facial = df['facewash']
vendas_sabonete = df['bathingsoap']

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.plot(meses, vendas_lavagem_facial, label='Lavagem Facial', color='b', marker='o')
ax1.set_title('Vendas de Lavagem Facial por Mês')
ax1.set_ylabel('Unidades Vendidas')
ax1.grid(linestyle='--')

ax2.plot(meses, vendas_sabonete, label='Sabonete', color='k', marker='o')
ax2.set_title('Vendas de Sabonetes por Mês')
ax2.set_ylabel('Unidades Vendidas')
ax2.set_xlabel('Número do Mês')
ax2.grid(linestyle='--')

plt.xticks(meses)
plt.tight_layout()


plt.show()
```

# **Exercício 10: Leia todos os dados de vendas de produtos e mostre-os usando o gráfico de pilha (stack plot)**





```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/company_sales_data - company_sales_data.csv")

months = df['month_number']

facecream_sales = df['facecream']
facewash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathingsoap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']

labels = ['Creme Facial', 'Lava-rosto', 'Pasta de Dente', 'Sabonete', 'Shampoo', 'Hidratante']

plt.figure(figsize=(12, 8))
plt.stackplot(months,
              facecream_sales, facewash_sales, toothpaste_sales,
              bathingsoap_sales, shampoo_sales, moisturizer_sales,
              labels=labels,
              colors=['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F', '#EDC948'])

plt.title('Vendas de Produtos por Mês (Gráfico de Pilha)', fontsize=16)
plt.xlabel('Número do Mês', fontsize=12)
plt.ylabel('Unidades Vendidas', fontsize=12)

plt.legend(loc='upper left')

plt.xticks(months)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
